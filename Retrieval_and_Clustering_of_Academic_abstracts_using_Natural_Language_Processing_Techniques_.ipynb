```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
stopwords = STOP_WORDS
import string
import en_core_sci_lg

# โหลดโมเดลภาษา
parser = en_core_sci_lg.load()
parser.max_length = 7000000  # กำหนดขนาดสูงสุดของข้อความที่สามารถประมวลผล

# กำหนด punctuation
punctuations = string.punctuation

def spacy_tokenizer(sentence):
    ''' Function to preprocess text of scientific papers
        (e.g Removing Stopword and punctuation)'''

    # ใช้ spacy tokenizer ในการแยกคำและทำ lemmatization
    mytokens = parser(sentence)

    # เปลี่ยนคำให้เป็น lowercase และ lemmatize (ไม่ใช้ pronouns)
    mytokens = [word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in mytokens]

    # กรอง stopwords และ punctuation
    mytokens = [word for word in mytokens if word not in stopwords and word not in punctuations]

    # รวมคำที่ได้กลับเป็นประโยค
    mytokens = " ".join([i for i in mytokens])
    return mytokens
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from tqdm import tqdm  # ✅ Import tqdm
tqdm.pandas()  # ✅ เปิดใช้งาน tqdm สำหรับ pandas
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix_subset = tfidf_vectorizer.fit_transform(df["processed_text"])
```

```python
cosine_sim_matrix = cosine_similarity(tfidf_matrix_subset, tfidf_matrix_subset)
```

```python
def search_best_match(query, tfidf_vectorizer, tfidf_matrix, df_subset):
    """
    ค้นหา Abstract ที่คล้ายกับคีย์เวิร์ดมากที่สุด (Top 1)
    """
    # 1. แปลงคีย์เวิร์ดเป็นเวกเตอร์ TF-IDF
    query_vec = tfidf_vectorizer.transform([query])

    # 2. คำนวณความคล้ายคลึงระหว่าง Query กับทุก Abstract
    cosine_sim = cosine_similarity(query_vec, tfidf_matrix)

    # 3. หาค่า Abstract ที่คล้ายที่สุด (อันดับ 1)
    best_idx = cosine_sim[0].argmax()
    best_score = cosine_sim[0][best_idx]

    # 4. แสดงผลลัพธ์ Abstract ที่ดีที่สุด
    print(f"\n🔍 **Best Match for:** \"{query}\"")
    print(f"- [{best_idx}] {df_subset.iloc[best_idx]['processed_text'][:150]}... (Score: {best_score:.4f})")

def search_multiple_queries(query_list, tfidf_vectorizer, tfidf_matrix, df_subset):
    """
    ค้นหา Abstract สำหรับรายการคำค้นหลายรายการ (Top 1 แต่ละคำ)
    """
    for i, query in enumerate(query_list):
        print(f"\n===== [Search {i+1}] =====")
        search_best_match(query, tfidf_vectorizer, tfidf_matrix, df_subset)
```

```python
import gensim
from gensim.models import Word2Vec
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# แปลง processed_text ให้เป็น list of tokens (แยกคำ)
sentences = [text.split() for text in df["processed_text"].dropna()]
```

```python
# 1. สร้าง CBOW model
model_cbow = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=0       # sg=0 แปลว่า CBOW
)

# 2. สร้าง Skip-gram model
model_sg = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1       # sg=1 แปลว่า Skip-gram
)
```

```python
import numpy as np

# ฟังก์ชันเดิมที่คุณมี สำหรับเฉลี่ยเวกเตอร์คำเป็นเวกเตอร์ประโยค
def get_sentence_vector(sentence, model):
    words = sentence.split()
    vecs = [model.wv[w] for w in words if w in model.wv]
    if not vecs:
        return np.zeros(model.vector_size)
    return np.mean(vecs, axis=0)

# สร้าง matrix ของ abstracts สำหรับแต่ละโมเดล
abstract_vectors_cbow = np.array([
    get_sentence_vector(txt, model_cbow)
    for txt in df["processed_text"].dropna()
])

abstract_vectors_sg = np.array([
    get_sentence_vector(txt, model_sg)
    for txt in df["processed_text"].dropna()
])
```

```python
def search_best_match_w2v(query, model, abstract_vectors, df_subset):
    # embedding
    query_vec = get_sentence_vector(query, model).reshape(1, -1)
    # คำนวณ cosine similarity
    cosine_sim = cosine_similarity(query_vec, abstract_vectors)
    idx   = cosine_sim[0].argmax()
    score = cosine_sim[0][idx]
    print(f"- [{idx}] {df_subset.iloc[idx]['processed_text'][:150]}... (Score: {score:.4f})")
```

```python
keywords = [
"performance prediction",
"tessellation techniques",
"research methods",
"delay analysis",
"dynamic reconfiguration",
"collaboration tools",
"database systems",
"real-time systems",
"generic programming",
"resource allocation",
"security risks",
"branching processes",
"predictive modeling",
"morphological analysis",
"routing",
"optimization",
"bandwidth management",
"system monitoring",
"sensitivity analysis",
"motion modeling"
]

# เตรียมลิสต์ของ (ชื่อโมเดล, โมเดล, เวกเตอร์ abstracts)
models = [
    ("CBOW",      model_cbow, abstract_vectors_cbow),
    ("Skip-gram", model_sg,   abstract_vectors_sg)
]

# วนรันทีละโมเดล ทีละคำค้น
for model_name, model, abstract_vecs in models:
    print(f"\n===== Results using {model_name} =====")
    for kw in keywords:
        print(f"\n>> Query: \"{kw}\"")
        search_best_match_w2v(kw, model, abstract_vecs, df)
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 1. เตรียม TaggedDocument
#    เอา processed_text แต่ละแถวเป็นเอกสาร และแท็กด้วย index
tagged_docs = [
    TaggedDocument(words=text.split(), tags=[i])
    for i, text in enumerate(df["processed_text"].dropna())
]
```

```python
# 2. สร้างสองโมเดล Doc2Vec
model_dm = Doc2Vec(
    documents=tagged_docs,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    epochs=20,
    dm=1        # dm=1 → Distributed Memory (เหมือน CBOW)
)

model_dbow = Doc2Vec(
    documents=tagged_docs,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    epochs=20,
    dm=0        # dm=0 → Distributed Bag of Words (เหมือน Skip-gram)
)
```

```python
# 4. ฟังก์ชันค้นหาด้วย Doc2Vec
def search_best_match_d2v(query, model, abstract_vectors, df_subset):
    # infer vector จาก query
    qv = model.infer_vector(query.split()).reshape(1, -1)
    # คำนวณ cosine similarity
    sims = cosine_similarity(qv, abstract_vectors)[0]
    idx   = sims.argmax()
    score = sims[idx]
    print(f"- [{idx}] {df_subset.iloc[idx]['processed_text'][:150]}... (Score: {score:.4f})")
    # สร้างเวกเตอร์ของ abstract ทั้งหมดจาก model_dm
abstract_vectors_dm = [model_dm.infer_vector(doc.words) for doc in tagged_docs]

# สร้างเวกเตอร์ของ abstract ทั้งหมดจาก model_dbow
abstract_vectors_dbow = [model_dbow.infer_vector(doc.words) for doc in tagged_docs]
```

```python
# 5. รันวนทั้งสองโมเดลกับ keywords
keywords = [
"performance prediction",
"tessellation techniques",
"research methods",
"delay analysis",
"dynamic reconfiguration",
"collaboration tools",
"database systems",
"real-time systems",
"generic programming",
"resource allocation",
"security risks",
"branching processes",
"predictive modeling",
"morphological analysis",
"routing",
"optimization",
"bandwidth management",
"system monitoring",
"sensitivity analysis",
"motion modeling"
]

models = [
    ("DM",   model_dm,   abstract_vectors_dm),
    ("DBOW", model_dbow, abstract_vectors_dbow)
]

for name, m, vecs in models:
    print(f"\n===== Results using {name} =====")
    for kw in keywords:
        print(f"\n>> Query: \"{kw}\"")
        search_best_match_d2v(kw, m, vecs, df)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
# 2) เตรียมเอกสาร (abstracts) และสร้าง embeddings
docs = df["processed_text"].dropna().tolist()
abstract_embeddings = model.encode(docs, convert_to_tensor=False)
```

```python
def search_best_match_bert(query, model, abstract_embeddings, df):
    # สร้าง embedding ของ query
    q_emb = model.encode([query], convert_to_tensor=False)[0]
    # คำนวณ cosine similarity กับทุก document
    sims = cosine_similarity([q_emb], abstract_embeddings)[0]
    # เลือกดัชนีที่คะแนนสูงสุด
    idx = sims.argmax()
    score = sims[idx]
    # พิมพ์ผลลัพธ์ตัวเดียว
    print(f"\n🔍 Best Match for \"{query}\":")
    print(f"- [{idx}] {df.iloc[idx]['processed_text'][:150]}... (Score: {score:.4f})")
```

```python
# 2. เตรียมลิสต์ 20 คำค้น
queries = [
"performance prediction",
"tessellation techniques",
"research methods",
"delay analysis",
"dynamic reconfiguration",
"collaboration tools",
"database systems",
"real-time systems",
"generic programming",
"resource allocation",
"security risks",
"branching processes",
"predictive modeling",
"morphological analysis",
"routing",
"optimization",
"bandwidth management",
"system monitoring",
"sensitivity analysis",
"motion modeling"
]
# 3) วนลูปเรียกใช้กับ df
for kw in queries:
    search_best_match_bert(kw, model, abstract_embeddings, df)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# สมมติ abstract_embeddings = model.encode(docs, convert_to_tensor=False)

# 1) ย่อมิติลง 2D
pca = PCA(n_components=2, random_state=42)
X2d = pca.fit_transform(abstract_embeddings)

# 2) เตรียม subplots: 2 แถว × 3 คอลัมน์ = 6 กราฟ สำหรับ k=2..7
Ks = range(2, 8)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for ax, k in zip(axes.flatten(), Ks):
    # 3) รัน KMeans
    labels = KMeans(n_clusters=k, random_state=42).fit_predict(abstract_embeddings)
    # 4) Plot
    ax.scatter(X2d[:, 0], X2d[:, 1], c=labels)
    ax.set_title(f'k = {k}')
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')

fig.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA


# 1) ลดมิติด้วย PCA (ลดเหลือ 2 มิติ)
pca = PCA(n_components=2, random_state=42)
embeddings_reduced = pca.fit_transform(abstract_embeddings)

# 2) วนลูปทำ clustering และคำนวณ Silhouette สำหรับ k = 2..7
for k in range(2, 8):
    # รัน KMeans บนข้อมูลที่ลดมิติแล้ว
    labels = KMeans(n_clusters=k, random_state=42).fit_predict(embeddings_reduced)

    # คำนวณ Silhouette score บนมิติที่ลดแล้ว
    score = silhouette_score(embeddings_reduced, labels)
    print(f"\n=== K = {k} clusters (silhouette = {score:.4f}) ===")

    # ดูขนาดแต่ละคลัสเตอร์
    counts = np.bincount(labels)
    for cluster_id, size in enumerate(counts):
        print(f" Cluster {cluster_id}: size = {size}")

    # แสดงตัวอย่างข้อความจากแต่ละคลัสเตอร์ 3 รายการ
    for cluster_id in range(k):
        idxs = np.where(labels == cluster_id)[0][:3]
        for i, idx in enumerate(idxs):
            snippet = docs[idx][:80].replace("\n", " ")
            print(f"  → Example C{cluster_id}_{i+1} [Index {idx}]: {snippet}...")
```
